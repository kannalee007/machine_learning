```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
m = 100
x = 6 * np.random.rand(m,1)-2.5
```

```python
y = 3*x**2 + 1.3*x + 4.5 #ax^2+bx+c=0 (quadratic equation)
```

```python
plt.plot(x,y, "b.")
plt.xlabel("$x$", fontsize = 13)
plt.ylabel("$y$", rotation = 0, fontsize = 13)
plt.axis([-3,3,0,40])
plt.show()
```

```python
#perfect quadratic data
```

```python
m = 100
x = 6 * np.random.rand(m,1)-2.5
y = 3*x**2 + 1.3*x + 4.5 +np.random.rand(100,132)
```

```python
plt.plot(x,y, "b.")
plt.xlabel("$x$", fontsize = 13)
plt.ylabel("$y$", rotation = 0, fontsize = 13)
plt.axis([-3,3,0,35])
plt.show()
```

```python
#not perfect like the first one
```

```python
feat = PolynomialFeatures(degree = 2, include_bias = False)
acess = feat.fit_transform(x)
acess
```

```python
linear_reg = LinearRegression()
linear_reg.fit(acess, y)
```

```python
linear_reg.intercept_, linear_reg.coef_
```
