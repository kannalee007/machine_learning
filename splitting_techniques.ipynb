```python
#ADVANCED STRATIFIED SPLIT (Three-Way Split Logic)
iris = load_iris()
x = pd.DataFrame(iris.data)
y = pd.Series(iris.target)
strat_shuffle = StratifiedShuffleSplit(n_splits=1, test_size=0.125, random_state=42)  #creating instance for StratifiedShuffleSplit class.
result = strat_shuffle.split(x,y)
result
for train_indices, test_indices in result:
    train_set = x.loc[train_indices]
    test_set = y.loc[test_indices]
print(test_set)
print(train_set)
```
