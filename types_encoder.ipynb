```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
y_original = iris.target.reshape(-1, 1) 

encoder = OrdinalEncoder()
y_encoded_ordinal = encoder.fit_transform(y_original)
print(y_encoded_ordinal)
```

```python
iris = load_iris()
# Reshape the target array to be 2D as required by the encoder
y_original = iris.target.reshape(-1, 1) 

# 1. Instantiate the encoder
# handle_unknown='ignore' prevents errors if unknown categories appear later
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# 2. Fit and transform the data
y_encoded_onehot = encoder.fit_transform(y_original)
print(y_encoded_onehot)
```
